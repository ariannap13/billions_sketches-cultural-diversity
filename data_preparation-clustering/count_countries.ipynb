```python
import os
import pandas as pd
```

```python
dict_counts = {}

# Loop over all country_counts CSV files in the folder
for file in os.listdir("../data/country_counts/"):
    if file.endswith("../data/country_counts/country_counts.csv"):
        df = pd.read_csv(file)

        df.rename(columns={"Unnamed: 0": "countrycode", "countrycode": "count"}, inplace=True)
        
        # Sum counts per country in the current file
        country_sums = df.groupby("countrycode")["count"].sum()
        
        # Add to the cumulative dictionary
        for country, count in country_sums.items():
            if country not in dict_counts:
                dict_counts[country] = count
            else:
                dict_counts[country] += count
```

```python
df_dict = pd.DataFrame(list(dict_counts.items()), columns=["country", "count"])
```

```python
df_dict["percentage"] = (df_dict["count"]/sum(df_dict["count"])).round(5)
```

```python
df_dict.sort_values(by="count", ascending=False).head(100)
```

```python
df_dict.sort_values(by="count", ascending=False)[["country", "percentage"]].head(100).to_csv("./table_counts.csv", index=False)
```
